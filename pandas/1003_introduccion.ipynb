#### Opción 2: Agregar una columna la cual contiene el resultado de una operaciones aritméticas en n filas

```python
# De esta forma calculo el Nuevo Total Facturado el cual debe ser igual a la columna Total Facturado

# Previamente se detecta gracias a la funcion .dtypes que las columnas "Precio", "Cantidad" y "Tot % bonif." son de tipo object, para este cálculo necesitamos que sean de tipo float.
# Procedemos a su conversion

if (varDataFrame['Precio'].dtypes == object):
    varDataFrame['Precio'] = varDataFrame['Precio'].str.replace('.', '').str.replace(',', '.').astype(float)

if (varDataFrame['Cantidad'].dtypes == object):
    varDataFrame['Cantidad'] = varDataFrame['Cantidad'].str.replace('.', '').str.replace(',', '.').astype(float)

if (varDataFrame['Tot % bonif.'].dtypes == object):
    varDataFrame['Tot % bonif.'] = varDataFrame['Tot % bonif.'].str.replace('.', '').str.replace(',', '.').astype(float)

if (varDataFrame['Total Facturado'].dtypes == object):
    varDataFrame['Total Facturado'] = varDataFrame['Total Facturado'].str.replace('.', '').str.replace(',', '.').astype(float)

varDataFrame.dtypes

varDataFrame['Nuevo Total'] = (varDataFrame['Precio'] * varDataFrame['Cantidad'])-(varDataFrame['Precio'] * varDataFrame['Cantidad'] * varDataFrame['Tot % bonif.'])

#varDataFrame['Nuevo Total']!='NaN'
#varDataFrame.info()

varDataFrame[['CAJA', 'SUCURSAL', 'Fecha de Venta', 'Precio', 'Cantidad', 'Tot % bonif.', 'Total Facturado', 'Vendedor', 'Nuevo Total' ]].head(100)
```
