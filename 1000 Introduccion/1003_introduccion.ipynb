```python
# Paso 1: Importamos librerias 
import pandas
import numpy
```

#### Paso 1: Cargamos el DataFrame en memoria a partir de un archivo csv

```python
# vamos a leer el contenido del archivo csv, para que funcione es importante que el archivo csv este en el mismo directorio que esta notebook
varDataFrame = pandas.read_csv('BTC-USD.csv')
```

### Transformaciones: Agregar una nueva columna

#### Opción 1: Agregar una nueva columna con un valor constante

```python
# De esta forma agrego una nueva columna llamada "Máximo Constante Manual" con el valor definido por el usuario 66000
varDataFrame['Maximo Constante Manual'] = 66000
varDataFrame
```

```python
# De esta forma agrego una nueva columna llamada "Máximo Constante Automatico" con el valor máximo de la columna High
varDataFrame['Maximo Constante Automatico'] = varDataFrame['High'].max()
varDataFrame
```

#### Opción 2: Sumar el contenido de n filas

```python
# De esta forma sumo el contenido de dos columnas para cda fila y lo almaceno en una nueva columna llamada Suma open y Close
varDataFrame['Suma Open y Close'] = varDataFrame['Open'] + varDataFrame['Close']
varDataFrame
```

```python
# De esta forma calculo el promedio entre la cotización de apertura y cierre
varDataFrame['Promedio'] = ( varDataFrame['Open'] + varDataFrame['Close'] ) /2
varDataFrame
```
