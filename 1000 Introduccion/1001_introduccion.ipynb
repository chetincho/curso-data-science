```python
# Paso 1: Importamos librerias 
import pandas
import numpy
```

### Opciones para mostrar un dataframe en pantalla

#### Paso 1: Cargamos el DataFrame en memoria a partir de un archivo csv

```python
# vamos a leer el contenido del archivo csv, para que funcione es importante que el archivo csv este en el mismo directorio que esta notebook
varDataFrame = pandas.read_csv('BTC-USD.csv')
```

#### Opción 1: Vista predeterminada de un DataFrame

```python
# De esta forma vemos un resumen con las primeras 5 filas y las últimas 5 filas de nuestro DataFrame
varDataFrame
```

#### Opción 2: Mostrar solo las primeras 5 filas del DataFrame

```python
# Para ello utilizamos el método head()
varDataFrame.head()
```

#### Opción 3: Mostrar solo las últimas 5 filas del DataFrame

```python
# Para ello utilizamos el método tail()
varDataFrame.tail()
```

#### Opción 4: Mostrar n filas del DataFrame

```python
# Para ello solo debemos especificar la cantidad de filas a mostrar dentro del método head() o tail() dependiendo si queremos mostrar las "n" primeras filas o las "n" ultimas filas
varDataFrame.head(8)
```

### Opción 5: Mostrar solo determinadas columnas del DataFrame

```python
# Para lograrlo solo tenes que colocar entre corchetes y comillas el nombre de la columna a consultar
varDataFrame['Date'].head(2)
# En este caso usamos una combinación con el método head para mostrar solo dos filas de la columna fecha
```

```python
# Para ver varias columnas solo debemos abrir nuevamente corchetes y luego separar con coma los nombres de las columnas a filtrar, por último debemos cerrar corchetes
varDataFrame[ ['Date', 'High', 'Low'] ].head(4)
```

### Opción 6: Mostrar n filas que cumplan con determinada condición

```python
# Muestro solo aquellas filas que cumplan con la condición High mayor o igual que 66000
filtrado_maximos = varDataFrame[varDataFrame['High'] >= 66000]
filtrado_maximos
```

```python
# Muestro solo aquellas filas que cumplen con las dis condiciones definidas
filtrado_n_condiciones = varDataFrame[
                                        (varDataFrame['High'] >= 66000) & 
                                        (varDataFrame['Close'] >= 67000)
                                     ]
filtrado_n_condiciones
```

### Opción 7: Mostrar valores máximos y mínimos de una determinada columna

```python
max = varDataFrame['High'].max()
min = varDataFrame['High'].min()

print(f'El valor máximo es: {max}')
print(f'El valor mínimo es: {min}')
```
