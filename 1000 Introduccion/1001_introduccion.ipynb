### Opciones para mostrar un dataframe en pantalla

#### Paso 1: Cargamos el DataFrame en memoria a partir de un archivo csv

```python
# vamos a leer el contenido del archivo csv, para que funcione es importante que el archivo csv este en el mismo directorio que esta notebook
varDataFrame = pandas.read_csv('data_historico_coti_btc.csv')
```

#### Opción 1: Vista predeterminada de un DataFrame

```python
# De esta forma vemos un resumen con las primeras 5 filas y las últimas 5 filas de nuestro DataFrame
varDataFrame
```

#### Opción 2: Mostrar solo las primeras 5 filas del DataFrame

```python
# Para ello utilizamos el método head()
varDataFrame.head()
```

#### Opción 3: Mostrar solo las últimas 5 filas del DataFrame

```python
# Para ello utilizamos el método tail()
varDataFrame.tail()
```

#### Opción 4: Mostrar n filas del DataFrame

```python
# Para ello solo debemos especificar la cantidad de filas a mostrar dentro del método head() o tail() dependiendo si queremos mostrar las "n" primeras filas o las "n" ultimas filas
varDataFrame.head(8)
```
